### Ejemplos Prácticos de Python en Agroindustria

Aquí tienes tres fragmentos de código básicos que ilustran tareas comunes en el sector:

#### 1. Cálculo de la Humedad en Base Húmeda

Este script calcula el porcentaje de humedad de una muestra, algo fundamental para el almacenamiento de granos.

```
# Datos de laboratorio
peso_inicial = 500.0  # gramos
peso_seco = 425.0     # gramos

# Cálculo de humedad
humedad = ((peso_inicial - peso_seco) / peso_inicial) * 100

print(f"El porcentaje de humedad de la muestra es: {humedad}%")
```

#### 2. Estimación de Rendimiento de Extracción

Ideal para plantas procesadoras de aceites o jugos, este código evalúa si la eficiencia de la máquina está dentro del rango esperado.

```
materia_prima = 1000  # kg de naranja
jugo_obtenido = 450   # litros

rendimiento = (jugo_obtenido / materia_prima) * 100

if rendimiento < 40:
    print("Alerta: Rendimiento bajo. Revisar presión de extractores.")
else:
    print(f"Rendimiento óptimo: {rendimiento}%")
```

#### 3. Conversión de Unidades de Temperatura (°C a °F)

Muchos manuales de maquinaria o normativas internacionales vienen en grados Fahrenheit. Este script facilita la conversión rápida para el control de pasteurización.

```
def celsius_a_fahrenheit(celsius):
    return (celsius * 9/5) + 32

temp_proceso = 72  # Grados Celsius para pasteurización
print(f"La temperatura de control debe ser: {celsius_a_fahrenheit(temp_proceso)} °F")
```
